```python
dataset = pd.read_csv("iris.data")
dataset
```

```python
X = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4].values

from sklearn.preprocessing import LabelEncoder
encoder =  LabelEncoder()
y1 = encoder.fit_transform(y)
Y = pd.get_dummies(y1).values

from sklearn.model_selection import train_test_split
X_train,X_test, y_train,y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
y1
```

```python
Y
```

```python
model = Sequential()

model.add(Dense(4,input_shape=(4,), activation='relu'))

model.add(Dense(3, activation='softmax'))
```

```python
model.compile(optimizer="Adam", loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.fit(X_train, y_train, batch_size=10, epochs=100)
```

```python
import numpy as np  # Ensure numpy is imported

y_pred = model.predict(X_test)  # Predict probabilities for test data
y_test_class = np.argmax(y_test, axis=1)  # Get actual class indices from one-hot encoded labels
y_pred_class = np.argmax(y_pred, axis=1)  # Get predicted class indices from probabilities
```

```python
from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(y_test_class,y_pred_class))
print(confusion_matrix(y_test_class,y_pred_class))
```
